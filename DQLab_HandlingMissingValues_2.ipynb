# Coba: Imputing data umur berdasarkan rata-rata dari masing-masing gender dan Menghapus Kolom Cabin.

## Impute Data
nilai mean dan median age dari masing gender hampir sama, sehingga umumnya kita bisa mengisi missing values dengan nilai mean dari age masing gender.

```python
dfCoba = df.copy()
```

```python
dfCoba.head()
```

```python
dfCoba[dfCoba['Age'].isnull() == True]
```

```python
dfCoba.isnull().sum()
```

```python
len(dfCoba[(dfCoba['Age'].isnull() == True) & (dfCoba['Sex'] == 'female')])
```

```python
len(dfCoba[(dfCoba['Age'].isnull() == True) & (dfCoba['Sex'] == 'male')])
```

```python
dfCoba['Age'].describe().T[['mean','50%']]
# Kolom Age Distribusinya Normal
```

```python
print(dfCoba[dfCoba['Sex'] == 'female']['Age'].mean())
print(dfCoba[dfCoba['Sex'] == 'female']['Age'].median())
```

```python
print(dfCoba[dfCoba['Sex'] == 'male']['Age'].mean())
print(dfCoba[dfCoba['Sex'] == 'male']['Age'].median())
```

```python
na_female = dfCoba[(dfCoba['Sex'] == 'female')]['Age']
na_male = dfCoba[(dfCoba['Sex'] == 'male')]['Age']

print(len(na_female))
print(len(na_male))
```

```python
na_female.fillna(np.round(dfCoba[dfCoba['Sex'] == 'female']['Age'].mean()), inplace=True)
na_male.fillna(np.round(dfCoba[dfCoba['Sex'] == 'male']['Age'].mean()), inplace=True)
```

```python
fillna = pd.concat([na_male, na_female]).sort_index()
```

```python
dfCoba['Age'] = fillna
```
